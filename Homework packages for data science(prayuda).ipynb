Memuat dan mencetak informasi data set dasar:

```python
import pandas as pd
df = pd.read_csv('ccaa_covid19_datos_isciii.csv')
df.info()
```

```python
#dataset tail
df.tail()
```

```python
# hitung berapa banyak missing values di setiap kolomnya
df.isnull().sum()
```

### Menulis program Python untuk mendapatkan jumlah kasus yang dikonfirmasi, dirawat di rumah sakit, dipindahkan ke ICU, meninggal dan sembuh.

```python
cases = df[['Fecha','Hospitalizados','UCI','Fallecidos','Recuperados']].groupby('Fecha')
cases.sum().sort_values('Fecha', ascending=False).head(1)
```

### Temukan jumlah terakhir dari kasus yang dikonfirmasi, meninggal dan dipulihkan oleh Komunitas:

```python
community = df[['CCAA','Casos','Fallecidos','Recuperados']].groupby('CCAA')
community.sum().sort_values('Casos', ascending=False)
```

### Temukan jumlah kasus terbaru yang dikonfirmasi, dirawat di rumah sakit, dipindahkan ke ICU, meninggal dan pulih dari COVID-19 di Navarra

```python
df.groupby('Fecha').sum().sort_values('Fecha', ascending=False).reset_index().head(1)
```

### Temukan kolom baru dengan kasus aktif:

```python
active = df.fillna(0)
active.loc[:,'Activos'] = active.Casos - active.Fallecidos - active.Recuperados
active.head()
```

### Temukan kapan kasus COVID-19 pertama dan kematian terjadi di Spanyol

```python
# cari kasus pertama (casos)
active[active['Casos'] > 0].sort_values('Fecha').head(1)
```

```python
# selediki kematian pertama ( falecidos )
active[active['Fallecidos'] > 0].sort_values('Fecha').head(1)
```

### Menemukan kapan kematian pertama kali terjadi di setiap Komunitas

```python
comm = active[active['Fallecidos'] > 0].sort_values(['Fecha', 'CCAA'])
comm.groupby('CCAA').first().sort_values(['Fecha','CCAA']).reset_index()
```
